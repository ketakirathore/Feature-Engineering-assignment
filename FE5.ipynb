```python
#IMPRTING PANDAS AS pd AND NUMPY AS np
import pandas as pd
import numpy as np
```

```python
#Employee Experience Tracker
```

```python
# 1. Create Raw Data
hr_data = {
    'Emp_Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Dept': ['Sales', 'IT', 'Sales', 'IT', 'HR'],
    'Salary': [70000, 95000, 65000, 110000, 60000],
    'Joined_Year': [2020, 2023, 2019, 2021, 2024]
}
hr = pd.DataFrame(hr_data)
```

```python
hr
```

```python
# 2. Feature Engineering
# Tenure in years
hr['Tenure_Years'] = 2026 - hr['Joined_Year']
```

```python
hr['Tenure_Years']
```

```python
# 3.Peer Comparison
```

```python
hr['Dept_Avg_Salary'] = hr.groupby('Dept')['Salary'].transform('mean')
hr['Salary_Ratio_to_Dept'] = hr['Salary'] / hr['Dept_Avg_Salary']
```

```python
hr['Salary_Ratio_to_Dept']
```

```python
#4. Seniority Flag based on tenure
hr['Senior_Level'] = np.where(hr['Tenure_Years'] > 4, 1, 0)
```

```python
hr['Senior_Level']
```
