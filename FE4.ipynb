```python
#IMPRTING PANDAS AS pd AND NUMPY AS np
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
#Food Delivery Time Predictor
```

```python
# 1. Create Raw Data
delivery_data = {
    'Order_ID': [101, 102, 103, 104, 105],
    'Timestamp': ['2026-02-14 18:30', '2026-02-15 11:15', '2026-02-16 19:45', '2026-02-14 13:00', '2026-02-17 08:20'],
    'Distance_KM': [2.5, 5.0, 1.2, 8.5, 3.1],
    'Cuisine': ['Pizza', 'Sushi', 'Burger', 'Pizza', 'Coffee']
}
delivery = pd.DataFrame(delivery_data)
```

```python
delivery
```

```python
# 2. Feature Engineering
delivery['Timestamp'] = pd.to_datetime(delivery['Timestamp'])
```

```python
delivery
```

```python
delivery['Timestamp']
```

```python
# Extracting Day Type
delivery['Is_Weekend'] = np.where(delivery['Timestamp'].dt.dayofweek >= 5, 1, 0)
```

```python
delivery['Is_Weekend']
```

```python
# Rush Hour Module (Dinner rush 6pm - 9pm)
delivery['Hour'] = delivery['Timestamp'].dt.hour
delivery['Is_Rush_Hour'] = np.where((delivery['Hour'] >= 18) & (delivery['Hour'] <= 21), 1, 0)
```

```python
delivery['Is_Rush_Hour']
```

```python
# Speed Category (Simple binning)
delivery['Distance_Category'] = pd.cut(delivery['Distance_KM'], bins=[0, 3, 7, 20], labels=['Short', 'Medium', 'Long'])
```
