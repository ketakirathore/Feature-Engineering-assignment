```python
# 1. Create Raw Data
car_data = {
    'Brand': ['Toyota', 'BMW', 'Ford', 'Tesla', 'Audi'],
    'Year': [2015, 2022, 2018, 2021, 2010],
    'Mileage': [85000, 15000, 45000, 22000, 150000],
    'Price_USD': [12000, 45000, 18000, 55000, 7000]
}
cars = pd.DataFrame(car_data)
```

```python
cars
```

```python
# 2. Feature Engineering
current_year = 2026
cars['Car_Age'] = current_year - cars['Year']
```

```python
cars
```

```python
# Usage Intensity: Miles driven per year
# We add 1 to Age to avoid division by zero for brand new cars
cars['Miles_Per_Year'] = cars['Mileage'] / (cars['Car_Age'] + 1)
```

```python
cars
```

```python
# Luxury Flag
luxury_brands = ['BMW', 'Tesla', 'Audi', 'Mercedes', 'Lexus']
cars['Is_Luxury'] = cars['Brand'].isin(luxury_brands).astype(int)
```
